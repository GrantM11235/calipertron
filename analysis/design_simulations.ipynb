```python
def filter(signal, sampling_frequency, cutoff_frequency, type="low"):
    b, a = scipy.signal.butter(6, cutoff_frequency, fs= sampling_frequency, btype=type, analog=False)    
    return scipy.signal.filtfilt(b, a, signal)
```

```python
pdm_num_samples = len(bsrr)
pdm_signal_frequency = pdm_frequency / pdm_num_samples

def bsrr_to_gpio_states(bsrr_values):
    num_samples = len(bsrr_values)
    gpio_states = np.zeros((num_samples, 16), dtype=int)
    
    for i, bsrr in enumerate(bsrr_values):
        set_bits = bsrr & 0xFFFF
        reset_bits = (bsrr >> 16) & 0xFFFF
        
        if i > 0:
            gpio_states[i] = gpio_states[i-1]
        
        for j in range(16):
            if set_bits & (1 << j):
                gpio_states[i, j] = 1
            elif reset_bits & (1 << j):
                gpio_states[i, j] = 0
    
    return gpio_states

# Generate GPIO state matrix
gpio_matrix = bsrr_to_gpio_states(bsrr)

# Print shape of the resulting matrix
print(f"GPIO matrix shape: {gpio_matrix.shape}")

# Optionally, visualize the first few rows of the matrix
print("\nFirst 5 rows of the GPIO matrix:")
print(gpio_matrix[:5])
```

# Generate our own PDM signal

```python
def generate_pdm_waveform(N):
    t = np.arange(N) / N
    analog_signal = np.sin(2 * np.pi * t)
    
    # Normalize the analog signal to range [0, 1]
    normalized_signal = (analog_signal + 1) / 2
    
    # Generate PDM signal
    pdm_signal = np.zeros(N, dtype=int)
    error = 0
    
    for i in range(N):
        if normalized_signal[i] > error:
            pdm_signal[i] = 1
            error += 1 - normalized_signal[i]
        else:
            pdm_signal[i] = 0
            error -= normalized_signal[i]
    
    return pdm_signal
```

# Phase measurement via cross correlation

The caliper's linear position is derived from the phase of the signal reflected back at us through the scale.
Do we need an analog low-pass filter to smooth the PDM pulses before we read it via the ADC, or does that all get taken care of in the correlation?

```python

# Measure phases via cross correlation. Use a sliding window so we can get many measurements of the phase and thus a sense of the "noise"
def measure_phases(samples, frequency):
    window_size = 150
    num_windows = len(samples) - window_size + 1
    phases = np.zeros(num_windows)
    t = np.arange(len(samples)) / frequency

    for i in range(num_windows):
        window = samples[i:i+window_size]

        sine_wave = np.sin(2 * np.pi * frequency * t[i:i+window_size])
        cosine_wave = np.cos(2 * np.pi * frequency * t[i:i+window_size])
        correlation_sine = np.sum(window * sine_wave)
        correlation_cosine = np.sum(window * cosine_wave)
        phases[i] = np.arctan2(correlation_sine, correlation_cosine)

    return phases
```

```python
def resample(samples, original_frequency, sampling_duration, dead_time_duration):
    """
    Resample a given set of samples using a continuous method.
    Works for both upsampling and downsampling.
    
    Parameters:
    samples (array-like): The original samples
    original_frequency (float): The frequency of the original samples in Hz
    sampling_duration (float): The duration of sampling for each new sample in seconds
    dead_time_duration (float): The duration of dead time between samples in seconds
    
    Returns:
    numpy.ndarray: The resampled signal
    """
    
    # Calculate the new target frequency
    target_frequency = 1 / (sampling_duration + dead_time_duration)
    
    # Calculate the total duration of the original signal
    total_duration = len(samples) / original_frequency
    
    # Calculate the number of points needed for the resampled signal
    num_points = int(total_duration * target_frequency)

    # Create an array to hold the resampled signal
    resampled = np.zeros(num_points)

    for i in range(num_points):
        # Calculate the start and end times for this sample
        start_time = i * (sampling_duration + dead_time_duration)
        end_time = start_time + sampling_duration
        
        # Convert times to fractional indices in the original sample array
        start_index = start_time * original_frequency
        end_index = end_time * original_frequency
        
        # Calculate the contribution of each original sample to the new sample
        index = np.arange(int(np.floor(start_index)), int(np.ceil(end_index)))
        weights = np.minimum(index + 1, end_index) - np.maximum(index, start_index)
        weights = np.maximum(weights, 0)  # Ensure non-negative weights
        
        # Handle edge cases
        if len(index) == 0:
            # We're between two samples, so interpolate
            left_index = int(np.floor(start_index))
            right_index = min(left_index + 1, len(samples) - 1)
            t = start_index - left_index
            resampled[i] = samples[left_index] * (1-t) + samples[right_index] * t
        else:
            # Compute weighted average
            valid_indices = index < len(samples)
            if np.any(valid_indices):
                resampled[i] = np.sum(weights[valid_indices] * samples[index[valid_indices]]) / np.sum(weights[valid_indices])
            else:
                resampled[i] = samples[-1]  # Use last sample if we're past the end

    return resampled
```

```python
adc_sample_cycles = 1.5
#adc_sample_cycles = 12.5
#adc_sample_cycles = 71.5
#adc_sample_cycles = 239.5
sampling_frequency = adc_frequency / (adc_sample_cycles + adc_sample_overhead_cycles)

transcribed_resampled = resample(transcribed, pdm_frequency, sampling_duration=adc_sample_cycles/adc_frequency, dead_time_duration=adc_sample_overhead_cycles/adc_frequency)
generated_resampled = resample(generated, pdm_frequency, sampling_duration=adc_sample_cycles/adc_frequency, dead_time_duration=adc_sample_overhead_cycles/adc_frequency)
```

```python
# Does this resample fn work as expected?
# Create a figure for comparing original and resampled signals
fig = go.Figure()

# Calculate time arrays for original and resampled signals
original_time = np.arange(len(transcribed)) / pdm_frequency
resampled_time = np.arange(len(transcribed_resampled)) / sampling_frequency

duration = 0.001
original_samples = int(duration * pdm_frequency)
resampled_samples = int(duration * sampling_frequency)

# Plot the original transcribed PWM signal for the first 100 ms
fig.add_trace(
    go.Scatter(x=original_time[:original_samples], y=transcribed[:original_samples], 
               mode='lines', name='Original Transcribed', line=dict(color='blue', shape='hv'))
)

# Plot the resampled transcribed PWM signal for the first 100 ms
fig.add_trace(
    go.Scatter(x=resampled_time[:resampled_samples], y=transcribed_resampled[:resampled_samples], 
               mode='markers', name='Resampled Transcribed', marker=dict(color='red', size=3))
)

# Update layout
fig.update_layout(
    height=600,
    width=800,
    title_text="Comparison of Original and Resampled Transcribed PDM Signal",
    showlegend=True,
    yaxis_title="Amplitude",
    xaxis_title="Time (seconds)",
    xaxis_range=[0, duration]
)

# Show the plot
fig.show()
```

```python
transcribed_resampled_phases = measure_phases(transcribed_resampled, sampling_frequency)
generated_resampled_phases = measure_phases(generated_resampled, sampling_frequency)
transcribed_resampled_phases_filtered = filter(transcribed_resampled_phases, sampling_frequency, cutoff_frequency=1000, type="low")
generated_resampled_phases_filtered = filter(generated_resampled_phases, sampling_frequency, cutoff_frequency=1000, type="low")
```
